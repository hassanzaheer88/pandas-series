```python
import numpy as np 
import pandas as pd 
```

#### Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

#### Removing Missing Data: 
It works on the basis of Rows.

```python
data2 = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df2 = pd.DataFrame(data2)
df2
```

```python
df2.dropna()              # each row has null value,so everything will be removed.
```

```python
df2
```

```python
# thresh = at least non-na values.
df2.dropna(thresh =3 )             # atleast 3 non na values in rows = remain present , else will be drop.
```

```python
df2.dropna(thresh=5)
```

```python
df2.dropna(thresh=1)
```

#### Filling Missing Data :
It is based on column wise.

```python
df2
```

```python
df2.fillna(0)
```

```python
df2                  # Not applying in original df2 , to apply in original , use inplace keyword.
```

```python
 # Filling of different values for different columns.
values = { 'A':0 , 'B':100, 'C':300, 'D':400 }
df2.fillna(value=values)
```

```python
df2
```

```python
df2.fillna( df2.mean() )
```
