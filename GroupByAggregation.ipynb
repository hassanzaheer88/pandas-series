```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales.
# groupby() creates an object and i cannot be seen. And it is only used for categorical data.

cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by Store and calculate the sum of Sales
store = df.groupby('Store')['Sales'].sum()
store
```

```python
# Group by multiple columns
# Group by Category and Store
cat = df.groupby( [ 'Category' , 'Store' ] )['Sales'].sum()
cat
```

### Aggregation :
combination of multiple things to produce single answer.(e.g mean of many elements)

```python
df['Sales'].mean()
```

```python
# mean, median, min, max , count, std
```

```python
df['Sales'].min()
```

```python
df['Sales'].max()
```

```python
df['Sales'].std()
```

```python
# agg() is used for multiple operations.
df['Sales'].agg([ 'mean', 'min' , 'max'])
```
