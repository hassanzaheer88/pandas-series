```python
## Series: A series is one dimentional labeled array capable of holding any data type.
# The axis labels are collectively called the index.
# labels are something like indexes.
```

```python
## Creating Series: 
# There are multiple ways to create a pandas series.
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b' , 'c' ]
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index = labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
