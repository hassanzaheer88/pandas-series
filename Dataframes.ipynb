```python
import numpy as np
import pandas as pd 
```

#### Creating a Dataframe

```python
# DataFrame From Dictionary.
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# DataFrame From List
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
```

```python
columns = [ "Name" , "Age" , "City" , "Salary" ]
df2 = pd.DataFrame(data_list, columns = columns)
df2
```

#### Selection And Indexing of Columns

```python
df2
```

```python
df2['Name']                 # Instead of 0,1...columns name, their actual columns names will be used.
```

```python
df2["City"]
```

```python
df2[ [ "Name" , "City" ] ]
```

#### Creating a new Column

```python
df2["Designation"] = [ "Doctor" , "Eng" , "Doctor" , "Eng" ]
```

```python
df2
```

#### Removing Columns

```python
df2.drop("Designation", axis = 1)       # BY default axis = 0 -> row, so df2.drop("Designation") -> Error , because "Designaion"  is at axis = 1
```

```python
df2
```

```python
# As After removing the "Desiignation" column, it is still showing in df2 , 
# so we will use keyword inplace.
df2.drop("Designation", axis = 1, inplace= True)
df2
```

```python
df2.drop([ "City" , "Salary" ] , axis = 1)
```

```python
df2
```

```python
df2.drop(0, axis = 0 )
```

#### Selecting Subsets of Rows and Columns

```python
df.loc[[0,1]][["City" , "Salary"]]             # df.loc[[0,1]] = select subset of rows then [["City" , "Salary"]] is used for columns. 
```

```python
df.loc[[2,3]][["Name","Age"]]
```

```python
df.loc[[2,3]][["City","Salary"]]
```

#### Conditional Selection:

```python
df2
```

```python
# People whose age is above 30.
df2[ df2["Age"] > 30 ]
```

```python
 # People whose age is above 30 and belongs to Paris.
df2[ (df2["Age"] > 30) & (df2["City"] == "Paris") ]
```
